```python
data.drop(['car_ID','CarName'],axis=1,inplace=True)
```

```python
obj_columns = ['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem']
for i in obj_columns:
  print("*"*30)
  print(i)
  print("*"*30)
  print(data[i].value_counts())
```

```python
from matplotlib.gridspec import GridSpec
from matplotlib import style
sns.set()
sns.set_palette("husl")

for i in obj_columns:

  fig=plt.figure(figsize=(10,3))
  grid=GridSpec(nrows=1,ncols=2,figure=fig)

  fig1=fig.add_subplot(grid[:,0])
  sns.countplot(x=i,data=data,ax=fig1,color='g',palette="husl")

  fig2=fig.add_subplot(grid[:,1])
  sns.stripplot(x=i,y='price',data=data,palette="husl").set_ylim()
```

```python
data['doornumber']=data['doornumber'].map({'two':2,'four':4})
data['cylindernumber']=data['cylindernumber'].map({'two':2,'three':3,'four':4,'five':5,'six':6,'eight':8,'twelve':12})
```

```python
categories = ['fueltype','aspiration','carbody','drivewheel','enginelocation','enginetype','fuelsystem']
from sklearn.preprocessing import LabelEncoder
# Encode the values using LabelEncoder
label_encoder = LabelEncoder()
for col in categories:
    data[col] = label_encoder.fit_transform(data[col])
data.head()
```

```python
data.dtypes
```

```python
correlation_matrix = data.corr()

plt.figure(figsize=(15, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Heatmap")
plt.show()
```

```python
def identify_correlated_features(corr_matrix, threshold=0.7):
    # Create a set to hold the correlated features
    correlated_features = set()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold:
                colname = corr_matrix.columns[i]
                correlated_features.add(colname)

    return correlated_features

threshold = 0.7
correlated_features = identify_correlated_features(correlation_matrix, threshold)
print(f"Features to remove: {correlated_features}")

for col in correlated_features:
  if col!='price':
    data = data.drop(columns=col)

print("Dataframe after removing correlated features:")
print(data.head())
```

```python
x=data.drop(['price'],axis=1)
x.head()
```

```python
y=data['price']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(x_train,y_train)
y_pred=lr_model.predict(x_test)
y_pred
```

```python
table=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
table
```

```python
plt.scatter(y_test, y_pred, color='red')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='green', linewidth=1)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values')
plt.show()
```

```python
print("slope of Regression Line:",lr_model.coef_)
print("constant:",lr_model.intercept_)
```

```python
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print("MAPE:",mean_absolute_percentage_error(y_test,y_pred))
print("R2:",r2_score(y_test,y_pred))
```
